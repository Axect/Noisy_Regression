```python
class NoisyData(Dataset):
    def __init__(self, df):
        x = df[:,0]
        y = df[:,1]
        self.x = torch.tensor(x, dtype=torch.float32).reshape((-1,1))
        self.y = torch.tensor(y, dtype=torch.float32).reshape((-1,1))

    def __len__(self):
        return self.x.shape[0]

    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]
```

```python
ds_train = NoisyData(df_train)
ds_test  = NoisyData(df_test)
```

## Normal MLP

```python
class MLP(L.LightningModule):
    def __init__(self, hparams):
        super().__init__()
        self.lr          = hparams['learning_rate']
        self.batch       = hparams['batch_size']
        self.hidden_size = hparams['hidden_size']
        self.num_layers  = hparams['num_layers']
        self.epochs      = hparams['epochs']
        self.dropout     = hparams['dropout']

        self.save_hyperparameters()

        net = []
        dropout_net = []
        dr = nn.Dropout(self.dropout)
        dropout_net.append(dr)
        net.extend([
            nn.Linear(1, self.hidden_size),
            nn.GELU(approximate='tanh'),
            dr,
        ])
        for _ in range(self.num_layers):
            dr = nn.Dropout(self.dropout)
            dropout_net.append(dr)
            net.extend([
                nn.Linear(self.hidden_size, self.hidden_size),
                nn.GELU(approximate='tanh'),
                dr
            ])
        net.append(nn.Linear(self.hidden_size, 1))
        self.net = nn.Sequential(*net)
        self.dr_net = dropout_net
        self.loss_fn = nn.MSELoss()

    def freeze(self):
        for dr in self.dr_net:
            dr.p = 0.0
    
    def forward(self, x):
        return self.net(x)

    def training_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = self.loss_fn(y, y_hat)
        self.log('train_loss', loss)
        return loss

    def validation_step(self, batch, batch_idx):
        x, y = batch
        y_hat = self(x)
        loss = self.loss_fn(y, y_hat)
        self.log('val_loss', loss)
        return loss

    def configure_optimizers(self):
        optimizer = optim.Adam(self.parameters(), lr=self.lr)
        return {
            "optimizer": optimizer,
            # "lr_scheduler": {
            #     "scheduler": OneCycleLR(
            #         optimizer,
            #         max_lr = self.lr,
            #         epochs = self.epochs,
            #         steps_per_epoch = len(ds_train) // self.batch + 1,
            #     ),
            #     "interval": "step",
            #     "monitor": "val_loss",
            #     "strict": True,
            # }
            "lr_scheduler": {
                "scheduler": PolynomialLR(
                    optimizer,
                    total_iters = self.epochs,
                    power = 2.0
                ),
                "interval": "epoch",
                "monitor": "val_loss",
                "strict": True,
            }
        }
```

```python
hparams = {
    'learning_rate': 1e-2,
    'batch_size': 32,
    'epochs': 500,
    'hidden_size': 16,
    'num_layers': 4,
    'dropout': 0.2,
}
```

```python
dl_train = DataLoader(ds_train, batch_size=hparams['batch_size'], shuffle=True)
dl_test  = DataLoader(ds_test, batch_size=hparams['batch_size'], shuffle=False)
```

```python
model = MLP(hparams)
model
```

```python
summary(model, input_size=(hparams['batch_size'], 1))
```

```python
wandb_logger = WandbLogger(
    project="Noisy_Regression"  
)
```

```python
trainer = Trainer(
    logger = wandb_logger,
    max_epochs = hparams['epochs'],
    devices=[0],
    accelerator='auto',
    enable_progress_bar=False,
    callbacks=[
        ModelCheckpoint(monitor='val_loss', mode='min', save_top_k=3),
        LearningRateMonitor(logging_interval="epoch"),
    ]
)
```

```python
trainer.fit(model, dl_train, dl_test)
```

```python
def decoder(model, dl):
    x_input = []
    y_input = []
    y_output = []

    for x, y in dl:
        x_input.extend(x.flatten())
        y_input.extend(y.flatten())

        x = x.to(model.device)
        y = y.to(model.device)

        y_hat = model(x)
        y_output.extend(y_hat.detach().cpu().numpy().flatten())

    return x_input, y_input, y_output
```

```python
freeze_model = MLP.load_from_checkpoint(candis[-1])
freeze_model.freeze()
x_input, y_input, y_output = decoder(freeze_model, dl_train)

X_data = np.column_stack([x_input, y_input])
X_data = X_data[X_data[:,0].argsort()]

X_hat = np.column_stack([x_input, y_output])
X_hat = X_hat[X_hat[:,0].argsort()]
```

## Monte Carlo Dropout

```python
def decoder_mcd(model, dl, S):
    x_input  = []
    y_input  = []
    y_output = []
    y_temp = []

    for x, y in dl:
        x_input.extend(x.flatten())
        y_input.extend(y.flatten())
    
        x = x.to(model.device)
        y = y.to(model.device)
    
        y_hat = model(x)
        y_temp.extend(y_hat.detach().cpu().numpy().flatten())

    y_output.append(y_temp)
    
    for _ in range(S-1):
        y_hat_temp = []
        for x, y in dl:
            x = x.to(model.device)
            y = y.to(model.device)
    
            y_hat = model(x)
            y_hat_temp.extend(y_hat.detach().cpu().numpy().flatten())
        y_output.append(y_hat_temp)

    return x_input, y_input, np.array(y_output)
```

```python
dl_train = DataLoader(ds_train, batch_size=hparams['batch_size'], shuffle=False)
```

```python
best_model.eval()
with torch.no_grad():
    x_input, y_input, y_output = decoder_mcd(best_model, dl_train, 1000)

X_data = np.column_stack([x_input, y_input])
X_data = X_data[X_data[:,0].argsort()]
```

```python
y_mean = y_output.mean(axis=0)
y_var  = y_output.var(axis=0)
tau    = 0.01 * (1. - best_model.dropout) / (2 * 800 * 0.99)
y_std  = np.sqrt(y_var) + 1. / tau
```

```python
X_hat = np.column_stack([x_input, y_mean, y_std])
X_hat = X_hat[X_hat[:,0].argsort()]
```

```python
plt.figure(figsize=(10,6), dpi=150)
plt.plot(X_data[:,0], X_data[:,1], '.', label='Data', alpha=0.5)
plt.plot(X_hat[:,0], X_hat[:,1], 'r-', label='NN', alpha=0.5)
plt.fill_between(X_hat[:,0], X_hat[:,1] - X_hat[:,2], X_hat[:,1] + X_hat[:,2], color='r', alpha=0.2)
plt.legend()
plt.show()
```
